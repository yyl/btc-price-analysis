```python
import pandas as pd
import datetime
import numpy as np
import scipy as sp
import os
import matplotlib.pyplot as plt
import matplotlib
from ggplot import geom_point
%matplotlib inline
# font = {'size'   : 18}
# matplotlib.rc('font', **font)
matplotlib.rcParams['figure.figsize'] = (12.0, 6.0)
```

```python
time_format = "%Y-%m-%dT%H:%M:%S"
data = pd.read_csv("./data.csv", names=['time', 'price'], index_col='time',
                   parse_dates=[0], date_parser=lambda x: datetime.datetime.strptime(x[:-6], time_format))
data.head()
```

Data points are roughly 10 seconds per point.

### Overall plot it out

```python
data.plot(rot=0)
```

## 2015 data

```python
df2015 = data['2015']
df2015.plot(rot=0)
```

### Moving average

Moving average is the average value of a consecutive subset of data. It is mostly used to smooth out the fluctuation and highlight the trend in general.

#### moving average per 10 minutes of 2015 data 

```python
pd.rolling_mean(df2015, 60).plot(rot=0, label='mov avg')
plt.legend()
```

Another moving metric of interest would be pairwise moving correlation. 

### Returns

Return is defined as the change in percentage over time:

$$r_t = \frac{p_t}{p_{t-1}} - 1$$

#### Daily return index

```python
df2015ohlc = df2015.resample('d', how='ohlc')
df2015ohlc.head()
```

```python
df2015ohlc['price']['close'].pct_change().plot(legend=True, label='daily return index')
```

What is the frequency of transaction for BTC? Due to the characteristics of BTC, it requires some time to complete a transaction. Daily might be a too-frequent estimation.
